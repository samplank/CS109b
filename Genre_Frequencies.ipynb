## Genre Frequencies

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
%matplotlib inline
matplotlib.rcParams.update({'font.size': 7})
%config InlineBackend.figure_format = 'png2x'
```

```python
genres = pd.read_pickle('./datasets/genres.pkl')
genre_names = np.load('./datasets/genre_names.npy')
predicted = np.load('./datasets/predictions/predictions_scratch.npy')

# Use genre names as column labels
predicted = pd.DataFrame(predicted)
```

```python
plt.figure(figsize=(10, 5), dpi=125)

genre_inds = np.arange(genres.shape[1])
plt.bar(genre_inds - 0.15, genres.sum(), width=0.3, color='#81C784', align='center', label='Original')
plt.bar(genre_inds + 0.15, predicted.sum(), width=0.3, color='#29B6F6', align='center', label='Predicted')
    
plt.xlabel('Genre')
plt.ylabel('Frequency')
plt.title('Genre Frequency (Original vs Predicted)')
plt.xticks(genre_inds, genre_names, rotation=90)

# Make legend box slighly less ugly
legend = plt.legend()
legend.get_frame().set_linewidth(0.5)

plt.tight_layout(); plt.show()
```
