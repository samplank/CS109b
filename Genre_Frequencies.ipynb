## Genre Frequencies

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
%matplotlib inline
matplotlib.rcParams.update({'font.size': 7})
%config InlineBackend.figure_format = 'png2x'
```

```python
genres = pd.read_pickle('./datasets/genres.pkl')
genre_names = np.load('./datasets/genre_names.npy')
predicted = np.load('./datasets/predictions/predictions_scratch.npy')

# Use genre names as column labels
genres.columns = genre_names
predicted = pd.DataFrame(predicted, columns=genre_names)
```

```python
fig, ax = plt.subplots(figsize=(10, 5), dpi=125)

index = np.arange(19)
bar_width = 0.35

rects1 = plt.bar(index, genres.sum(), bar_width,
                 color='#29B6F6',
                 label='True Labels')

rects2 = plt.bar(index + bar_width, predicted.sum(), bar_width,
                 color='#81C784',
                 label='Predictions')

plt.xlabel('Genre')
plt.ylabel('Frequency')
plt.title('Genre Frequency')
plt.xticks(index + bar_width / 2, genre_names, rotation=90)
plt.legend()

plt.tight_layout()
plt.show()
```
