## Genre Frequencies

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
%matplotlib inline
matplotlib.rcParams.update({'font.size': 7})
%config InlineBackend.figure_format = 'png2x'
```

```python
genres = pd.read_pickle('./datasets/genres.pkl')
genre_names = np.load('./datasets/genre_names.npy')
predicted = np.load('./datasets/predictions/predictions_scratch.npy')

# Use genre names as column labels
genres.columns = genre_names
predicted = pd.DataFrame(predicted, columns=genre_names)
```

```python
_, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5), dpi=125)

ax1.bar(range(genres.shape[1]), genres.sum(), align='center')
ax1.set_xlabel('Genre')
ax1.set_ylabel('Frequency')
ax1.set_title('Genre Frequency (Original)')
ax1.set_xticks(range(genres.shape[1]))
ax1.set_xticklabels(genres.columns, rotation=90)

ax2.bar(range(predicted.shape[1]), predicted.sum(), align='center')
ax2.set_xlabel('Genre')
ax2.set_ylabel('Frequency')
ax2.set_title('Genre Frequency (Predicted)')
ax2.set_xticks(range(predicted.shape[1]))
ax2.set_xticklabels(predicted.columns, rotation=90)

plt.tight_layout(); plt.show()
```
